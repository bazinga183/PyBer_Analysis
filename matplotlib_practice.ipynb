```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with the ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, marker = "*", color = "blue", linewidth = 2, label = "Boston")

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare ($)")

#Make the line wider.



# Set the x limit b/w Jan and June.
plt.xlim(0, 5)

#Set the y limit b/w 0 and 45.
plt.ylim(0, 45)

# Create a title.
plt.title("PyBer Fare by Month")

# Add the legend.
plt.legend()
```

```python
fig, ax = plt.subplots(figsize = (10, 5))
ax.plot(x_axis, y_axis, label = 'Boston', color = 'green', marker = 'x')
ax.set_xlim(0, 12)
ax.set_ylim(0, 45)
ax.set_title('PyBer Fare By Month')
ax.set_xlabel('Date')
ax.set_ylabel('Fare ($)')
ax.legend()
ax.grid()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plt
plt.bar(x_axis, y_axis)
```

```python
# Create the plot.
plt.bar(x_axis, y_axis, color = 'green', label = 'Boston')

# Create the labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare ($)")

# Create a title.
plt.title("PyBer Fare By Month")

# Add a legend.
plt.legend()
```

```python
# Create the plot.
plt.barh(x_axis, y_axis, color = 'green', label = 'Boston')
plt.gca().invert_yaxis()

# Create the labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare ($)")

# Create a title.
plt.title("PyBer Fare By Month")

# Add a legend.
plt.legend()
```

```python
fig, ax = plt.subplots(figsize = (10, 5))
ax.barh(x_axis, y_axis, color = "cyan", label = "Chicago")
ax.legend()
ax.invert_yaxis()
```

```python
# Create the scatter

plt.scatter(y_axis, x_axis, c = 'red', label = 'Chicago', s = [i * 3 for i in y_axis])
plt.xlabel("Fare ($)")
plt.ylabel("Date")
plt.title("PyBer Fare By Month")
plt.gca().invert_yaxis()
y_axis_larger = []
```

```python
# Create the plot
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, c = 'skyblue', s = [i * 5 for i in y_axis], alpha = .8, edgecolors = 'black', label = 'Boston')
ax.legend()
ax.set_title("PyBer Fare by Month")
ax.invert_yaxis()
ax.set_xlim(0, 50)
```

```python
plt.pie(y_axis, labels = x_axis)
plt.show()
```

```python
plt.subplots(figsize = (8, 8))
explode_values = (0, 0, 0, 0, 0, 0, .2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode = explode_values, labels = x_axis, autopct = "%.1f%%")
plt.show
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize = (8, 8))
plt.pie(y_axis, 
       explode = explode_values,
       colors = colors, 
       labels = x_axis,
       autopct = '%.1f%%')
plt.show()
```

```python
fig, ax = plt.subplots()
explode_values = (0, 0, 0.5, 0, 0, 0, 0.5, 0, 0, 0, 0, 0)
ax.pie(y_axis, labels = x_axis, autopct = '%.1f%%', pctdistance = .8, radius = 2, explode = explode_values, counterclock = False, shadow = True,
      startangle = 90, colors = colors)
plt.show()
```
